###  Yearly Sales of Anti-Inflammatory Drugs :-

```python
plt.figure(figsize = (6,6))
df_anti_inflamm_drugs = df.groupby(['Year'])['M01AB'].sum()

df_anti_inflamm_drugs.plot(kind = 'bar',color='#607c5c', edgecolor='red', width=0.5, linewidth=4)

plt.title("M01AB", fontdict = {'fontsize' : 11})
plt.xlabel("Year")
plt.ylabel("Total_Sales[Volume]")
plt.show()
```

### From the above we concluded that in the year 2016, we sold maximum no of Anti-inflammed Drugs followed by 2015, 2017.

### Monthly Sales Time Series :-

```python
### M01AB - Anti-inflammatory and antirheumatic products, non-steroids; Acetic acid derivatives and related substances:--
```

```python
plt.figure(figsize = (15, 5))

df_anti_inflam = df.groupby(['Year','Month'])['M01AB'].sum()
df_anti_inflam.plot(kind = 'line', color ='#607c5c', marker='.')

ticks = [0,12,24,36,48,60,72]
labels = ['2014','2015','2016','2017','2018','2019','2020']

plt.title("M01AB_Drugs", fontdict = {'fontsize': 12})
plt.xlabel("Montly Sales")
plt.ylabel("Sum of Sales(Volume)")
plt.xticks(ticks = ticks, labels = labels)
plt.show()
```

```python
df1 = df.drop(['datum','Year','Month','Weekday Name'], axis = 1)
df1.groupby('Hour', as_index = False).mean()
display(df1[0:5])
```

```python
time = df1['Hour']

items = ['M01AB','M01AE','N02BA','N02BE','N05B','N05C','R03','R06']
for item in items:
  plt.plot(time, df1[item], label = item)

plt.title("Sales by Product Sales")
plt.xlabel('Hour')
plt.ylabel('sales')
plt.legend()
plt.show()
```
